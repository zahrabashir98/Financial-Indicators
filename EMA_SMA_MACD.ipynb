## SMA and EMA comparison

```
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd 

time        = np.arange(0, 10, 0.1);
amplitude   = np.sin(time)
numpy.savetxt("foo.csv", amplitude, delimiter=",")

data = pd.read_csv("foo.csv") 
sma_20 = data.rolling(window=20).mean()
ema_20 = data.ewm(span=20).mean()
plt.plot(data, label="Close Data")
plt.plot(sma_20, label="SMA_20")
plt.plot(ema_20, label="EMA_20")
plt.legend()
```

## MACD Indicator Implementation

```
import numpy as np
import matplotlib.pyplot as plot

datafile = "BINANCE_BTCUSDT, 480.csv"
data = pd.read_csv(datafile)

close = data['close']
print(len(close))

close.index = pd.to_datetime(close.index)
new_data = close[0:250]
new_data.plot()

ema26 = new_data.ewm(span=26).mean()
ema12 = new_data.ewm(span=12).mean()
ema26.plot()
ema12.plot()
diff = ema12 - ema26
diff.plot()
```
